```python
import pandas as pd
import numpy as np
from random import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv("regLin.csv")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[['X']], df[['y']], train_size=0.75)
```

```python
linReg = linear_model.LinearRegression()
linReg.fit(X_train, Y_train)
```

```python
print linReg.intercept_, linReg.coef_
```

```python
pred = linReg.predict(X_train)
plt.scatter(X_train,Y_train, color='blue')
plt.plot(X_train, pred, color = 'red')
plt.show()
```

```python
def err(X,y,w,w0):
    return np.mean((w0 + w*X.values- y.values)**2)

w0 = np.linspace(0,6,len(X_test))
w1 = np.linspace(-2,3,len(X_test))

inter, c = linReg.intercept_, linReg.coef_
errpred= err(X_test, Y_test, c, inter) 

err0 = [err(X_test, Y_test, c, w0[i]) for i in range(0,len(w0))]
err1 = [err(X_test, Y_test, w1[i], inter) for i in range(0,len(w1))]
```

```python
plt.plot(w0,err0)
plt.scatter(inter, errpred, color='red')
plt.show()
```

```python
plt.plot(w1,err1)
plt.scatter(c, errpred, color='red')
plt.show()
```
