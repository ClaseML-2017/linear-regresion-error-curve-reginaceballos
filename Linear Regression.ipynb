# Linear Regression

```python
import pandas as pd
import numpy as np
from random import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv("regLin.csv")
```

## Train and test sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[['X']], df[['y']], train_size=0.75)
```

## Linear regression

```python
linReg = linear_model.LinearRegression()
linReg.fit(X_train, Y_train)
```

```python
inter, c = linReg.intercept_, linReg.coef_
print inter, c
```

## Data and regression plot

```python
pred = linReg.predict(X_train)
plt.scatter(X_train,Y_train, color='blue')
plt.plot(X_train, pred, color = 'red')
plt.show()
```

```python
def err(X,y,w,w0):
    return np.mean((w0 + w*X.values- y.values)**2)
```

```python
w0 = np.linspace(float(inter-20),float(inter+20),len(X_test))
w1 = np.linspace(float(c-2),float(c+2),len(X_test))

errpred= err(X_test, Y_test, c, inter) 

err0 = [err(X_test, Y_test, c, w0[i]) for i in range(0,len(w0))]
err1 = [err(X_test, Y_test, w1[i], inter) for i in range(0,len(w1))]
```

## Error plot for different values of the intercept

```python
plt.plot(w0,err0)
plt.scatter(inter, errpred, color='red')
plt.show()
```

## Error plot for different values of the slope

```python
plt.plot(w1,err1)
plt.scatter(c, errpred, color='red')
plt.show()
```

Es fcil ver con las grficas que el error se minimiza utilizando mnimos cuadrados. Adems, en la grfica de datos y regresin podemos ver que la tendencia de los datos es claramente la lnea de regresin. Por lo tanto, se puede decir que el modelo ajusta bien a los datos.
